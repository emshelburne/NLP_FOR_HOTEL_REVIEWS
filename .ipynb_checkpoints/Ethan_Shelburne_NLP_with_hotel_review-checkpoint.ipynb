# Introduction

Hotel Management Inc. is looking to better understand what aspects of a hotel stay contribute most to customer satisfaction. We employ a large dataset of guest reviews, including structured data about the stay and free-text fields for positive and negative comments. This project aims to identify the key drivers of positive sentiment, as reflected in the binary Reviewer_Score (1 for positive, 0 for negative). By combining text analysis with structured features, we seek to offer interpretable, data-driven insights that can help improve guest experiences.

The analysis begins with exploratory data analysis to assess the structure, distribution, and quality of the data. We then preprocess the review text using Scikit-learn's CountVectorizer() tool, separately transforming positive and negative comments and combining them with the numeric features. Two models are developed and compared: a logistic regression model for interpretability and a PCA-augmented decision tree for performance tuning. We evaluate both using accuracy, precision, recall, F1 score, and confusion matrices, with a focus on the most predictive words driving review sentiment. These insights can inform targeted improvements in service and operations.

### Helper Functions

````python
# Define function to examine dataframes

def examine_df(df):
   
    """
    Check basic info about a dataframe df
    """
    
    print(f"\n\nNumber of records is: {len(df)}\n")
    print(f"The columns are: {df.columns}\n")
    print(f"\n General info:")
    display(df.info())
    print(f"\n Statistical info:")
    display(df.describe())
    print(f"\n\nSample of records:")
    display(df.head(5))
````

````python
# Define function to generate correlation heatmap for numeric columns of dataframe

def gen_corr_heatmap(df):

    """
    Generate correlation heatmap for numeric columns of a dataframe with plotly
    """
    
    num_df = df.select_dtypes(include='number')  # Restrict to numeric columns
    corr_matrix = num_df.corr()  # Compute correlation matrix

    fig = px.imshow(
        corr_matrix,
        text_auto=".2f",  # Show only two decimal points
        color_continuous_scale='RdBu',
        aspect='auto',
        title='Correlation Heatmap of Numeric Columns'
    )

    fig.update_layout(
        title={'x': 0.5},
        width=1000,  # Increase figure width
        height=1000,  # Increase figure height
        font=dict(size=8),  # Smaller font
        xaxis=dict(tickangle=45),  # Slant x-axis labels
        yaxis=dict(tickangle=0),   # Keep y-axis labels horizontal
    )

    fig.show()
````

````python
# Define function to plot scatter plots comparing numeric columns in a dataframe

def scatter_matrix(df, sample_size=1000):

    """
    Generate scatter plots comparing pairs of numeric columns (with samples of records to avoid overplotting) 
    """
    
    num_df = df.select_dtypes(include='number').copy()
    sample_size = min(sample_size, len(num_df))
    num_df = num_df.sample(n=sample_size, random_state=2025)

    # Generate all unique (x, y) pairs
    cols = num_df.columns
    pairs = list(itertools.combinations(cols, 2))

    # Set up figure
    fig, axes = plt.subplots(5, 2, figsize=(20, 28))  # 5 rows × 2 cols = 10 subplots
    axes = axes.flatten()

    for i, (x, y) in enumerate(pairs):
        ax = axes[i]
        ax.scatter(num_df[x], num_df[y], s=5, alpha=0.5)
        ax.set_title(f'{x} vs {y}', fontsize=9)
        ax.tick_params(axis='both', which='major', labelsize=7)

    # Turn off any unused subplots (in case of fewer than 10 pairs)
    for j in range(len(pairs), len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
````

````python
# Define function to plot probabilities of target variable vs. some feature's distribution

def plot_pos_probability(df, feature, target='Reviewer_Score', sample_size=10000, bins=30):

    """
    Generate plot of probability of target variable vs. some feature's distribution
    """

    
    df = df[[feature, target]].dropna()
    if len(df) > sample_size:
        df = df.sample(n=sample_size, random_state=2025)

    # Bin the feature
    df['bin'], bin_edges = pd.cut(df[feature], bins=bins, retbins=True, labels=False)
    bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])

    # Group by bin and compute conditional probability
    prob_by_bin = df.groupby('bin')[target].agg(['mean', 'count']).reset_index()
    prob_by_bin['bin_center'] = bin_centers[prob_by_bin['bin']]

    # Filter out bins with very low count for stability
    prob_by_bin = prob_by_bin[prob_by_bin['count'] >= 10]

    # Plot
    plt.figure(figsize=(8, 5))
    plt.plot(prob_by_bin['bin_center'], prob_by_bin['mean'], marker='o')
    plt.ylim(-0.05, 1.05)
    plt.title(f'Estimated P(Reviewer_Score = 1 | {feature})')
    plt.xlabel(feature)
    plt.ylabel('Probability of Positive Review')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
````

## 1.2 Data Cleaning

In this subsection, we undergo some basic data cleaning to prepare for exploratory statistical analysis and modelling.

````python
# Check if there are any duplicated rows

has_duplicates = hotel_df.duplicated().any()
print(f"Duplicated Rows: {has_duplicates}")
````

````python
# Drop unnecessary columns

cols_to_drop = ['Unnamed: 0', 'Days_Since_Review', 'Day', 'Month', 'Year','Additional_Number_of_Scoring','Average_Score','Total_Number_of_Reviews']

hotel_df.drop(columns = cols_to_drop, inplace = True)
````

We chose to drop the following columns from the dataset:

1. Unnamed: 0: This is an index column from the original CSV file and does not carry any meaningful information.

2. Days_Since_Review, Day, Month, Year: These columns were used to construct a proper datetime column, so they are now redundant.

3. Additional_Number_of_Scoring: This refers to the number of guests who submitted a numerical score without leaving a written review. Since it doesn't reflect the experience of the reviewer providing the current review, it doesn't add value to our analysis.

4. Average_Score and Total_Number_of_Reviews: Both of these are hotel-level aggregates, independent of the guest’s specific stay. While they could improve model performance by capturing overall hotel quality, including them risks creating a feedback loop where well-reviewed hotels are predicted to continue receiving positive reviews. This may inflate accuracy but limits our ability to extract actionable insights about what aspects of a stay lead to guest satisfaction.

By removing these columns, we aim to focus the analysis on features that more directly reflect the individual guest experience.

````python
# Replace date column with a cyclical encoding of month to capture seasonal trends

# Extract month
hotel_df['Review_month'] = hotel_df['Date'].dt.month

# Encode month as cyclical feature
hotel_df['Review_month_sin'] = np.sin(2 * np.pi * hotel_df['Review_month'] / 12)
hotel_df['Review_month_cos'] = np.cos(2 * np.pi * hotel_df['Review_month'] / 12)

# Drop the original date and raw month columns
hotel_df.drop(columns=['Date', 'Review_month'], inplace=True)
````

To capture potential seasonal trends in review sentiment, we transformed the review date into a cyclical representation of the month. Since months follow a circular pattern (e.g., December is close to January), we used sine and cosine transformations to encode this cycle. This approach allows models to recognize seasonal patterns without treating months as linear or unordered categories. After encoding, we dropped the original date and raw month columns, as their information is now embedded in the cyclical features. 

We also excluded the day and year components, as they are unlikely to offer meaningful patterns for modelling or business insights. Individual days tend to introduce noise rather than signal, and the data only spans a little over two years, limiting the usefulness of the year variable for capturing long-term trends.

````python
# Add binary columns for when a positive or negative review is missing


hotel_df['No_Positive'] = (hotel_df['Positive_Review'] == 'No Positive').astype(int)
hotel_df['No_Negative'] = (hotel_df['Negative_Review'] == 'No Negative').astype(int)

# Replace 'No Positive' and 'No Negative' with empty strings
hotel_df.loc[hotel_df['Positive_Review'] == 'No Positive', 'Positive_Review'] = ''
hotel_df.loc[hotel_df['Negative_Review'] == 'No Negative', 'Negative_Review'] = ''
````

We next addressed how missing reviews are represented in the dataset. Instead of using null values, the data uses the strings "No Positive" and "No Negative" to indicate the absence of a review. To preserve this information, we created two binary indicator columns: No_Positive and No_Negative. These can help capture patterns in guest feedback behavior, such as choosing to only leave negative comments or no text at all. We then replaced the placeholder strings with empty text to prepare the review fields for text processing. This step is important because, if left unchanged, the model might incorrectly interpret the words “positive” or “negative” as meaningful review content, potentially distorting the results.

## 2.2 Define Text Cleaning and CountVectorizer Pipeline

In this subsection, we set up CountVectorizer() objects with built in text preprocessing and tokenizing. 

````python
# Define function to clean and process text then return list of tokens (unigrams and bigrams)

def clean_tokenizer(text):

    """
    Given a text object, execute basic text cleaning (lowercase, remove punctuation, and stop words) then return list of unigrams and bigrams
    """
    
    # Lowercase
    text = text.lower()
    
    # Remove other punctuation and digits
    text = re.sub(r"[^a-z\s]", "", text)
    
    # Tokenize by whitespace
    tokens = text.split()
    
    # Remove stopwords and single letter words
    tokens = [token for token in tokens if token not in ENGLISH_STOP_WORDS and len(token) != 1]

    # Generate bigrams
    bigrams = [f"{tokens[i]}_{tokens[i+1]}" for i in range(len(tokens)-1)]
    
    # Return combined list of unigrams and bigrams
    return tokens + bigrams
````

````python
# Define CountVectorizers for positive and negative reviews

# Define vectorizer for positive reviews
pos_vectorizer = CountVectorizer(
    tokenizer=clean_tokenizer,
    max_features=500,
    min_df=10
)

# Define vectorizer for negative reviews
neg_vectorizer = CountVectorizer(
    tokenizer=clean_tokenizer,
    max_features=500,
    min_df=10
)
````

For text processing, we follow a structured approach to transform the positive and negative review columns into usable numerical features for modelling. We define a custom tokenizer that performs basic cleaning: it converts text to lowercase, removes punctuation and digits, and filters out stopwords and single-letter tokens. To capture more context, we also include bigrams alongside unigrams, combining them into a single list of tokens for each review.

We use this tokenizer in two separate CountVectorizer instances: one for the positive reviews and one for the negative reviews. Each vectorizer is limited to the 500 most frequent tokens, excluding those that appear fewer than 10 times in the dataset. Processing the positive and negative reviews separately allows us to retain their distinct signals.

With our data fully prepared, we now move into the modelling stage.
````python
# Define function to evaluate classification model

def evaluate_class_model(y_true, y_pred,dataset_name = ""):
    
    """
    Evaluate logistic regression model on test set using accuracy, precision, recall, and F1 scores
    """
    
    print(f"Evaluation on {dataset_name} set:")
    print(f"Accuracy  : {accuracy_score(y_true, y_pred):.3f}")
    print(f"Precision : {precision_score(y_true, y_pred):.3f}")
    print(f"Recall    : {recall_score(y_true, y_pred):.3f}")
    print(f"F1 Score  : {f1_score(y_true, y_pred):.3f}")
    print("\nFull Classification Report:")
    print(classification_report(y_true, y_pred))
    print()
````

````python
# Define function to plot influential coefficients in a logistic regression model on a bar graph

def plot_logreg_coefficients(coef_df, title="Top Influential Features"):
    
    """
    Plot logistic regression coefficients from a DataFrame on a bar graph
    """
    
    # Sort so that largest coefficients appear at the top
    plot_df = coef_df.sort_values(by='coefficient', ascending=True)

    # Assign custom colors based on sign of coefficient
    plot_df['color'] = plot_df['coefficient'].apply(
        lambda x: 'rgba(31, 119, 180, 0.8)' if x > 0 else 'rgba(214, 39, 40, 0.8)'
    )

    fig = go.Figure(
        data=go.Bar(
            x=plot_df['coefficient'],
            y=plot_df['feature'],
            orientation='h', # Horizontal orientation
            marker_color=plot_df['color'],
        )
    )

    fig.update_layout(
        title=title,
        xaxis_title='Coefficient Value',
        yaxis_title='Feature',
        height=600,
        yaxis=dict(tickfont=dict(size=10)),
        template='plotly_white'
    )

    fig.update_layout(yaxis=dict(tickfont=dict(size=10)), height=600)
    fig.show()
````

To build our logistic regression model, we set up a simple pipeline using StandardScaler() followed by LogisticRegression(). The scaler standardizes numeric features to have zero mean and unit variance, which helps the model converge more efficiently and avoids giving undue weight to features on larger scales. We use LogisticRegression() with penalty='l2' for regularization to prevent overfitting, and set C=1 to apply default regularization strength. We also set max_iter=1000 to allow sufficient iterations for convergence and use solver='liblinear', which is well-suited for binary classification models with regularization.

While hyperparameter tuning would typically follow, the size of our dataset (over 1,000 features and 345,000 records) makes exhaustive cross-validation computationally infeasible within our constraints. For this reason, we proceed these parameters, focusing instead on evaluating performance and interpretability in downstream analysis.

````python
# Fit model to train set

# Start timer
start_time = time.time()

log_reg_pipeline.fit(X_train_final, y_train)

end_time = time.time()
run_time = end_time - start_time

print(f"\n Logistic Regression Classifier completed fitting data in {run_time:.2f} seconds.")
````

## 3.2 Decision Tree Classifier

In this subsection, we build a decision tree classifier using a pipeline that includes PCA for dimensionality reduction, followed by hyperparameter optimization to evaluate its performance and compare it with our logistic regression model.

````python
# Define decision tree pipeline

dt_pipeline = Pipeline([
    ('scaler', StandardScaler()),     # Scale
    ('pca', PCA(n_components=20)),    # Apply PCA
    ('clf', DecisionTreeClassifier(random_state=2025)) # DT Classifier
])
````

To set up our decision tree pipeline, we begin by applying StandardScaler() to normalize the feature values. While decision trees are not sensitive to feature scaling, this step ensures compatibility with the PCA step that follows. We then reduce dimensionality using PCA(n_components=20), which helps simplify the feature space and speed up training, especially given the high number of input features. Finally, we apply a DecisionTreeClassifier() with a fixed random seed for reproducibility. 

````python
# Create hyperparamter grid

param_grid = {
    'clf__max_depth': [5, 10],              # controls tree complexity
    'clf__min_samples_leaf': [1, 5],        # limits overfitting
    'clf__min_samples_split': [2, 5]        # regularizes further
}
````

To optimize our decision tree model, we define a grid of hyperparameters to search over during cross-validation. The parameter max_depth controls the maximum depth of the tree, with smaller values helping to prevent overly complex trees that may overfit the training data. We test values of 5 and 10 to balance depth and generalization. The min_samples_leaf parameter sets the minimum number of samples required at a leaf node, which helps regularize the model by preventing splits that result in very small, possibly noisy leaves. We include values of 1 and 5 to evaluate both flexible and more conservative splits. Finally, min_samples_split controls the minimum number of samples required to split an internal node. By testing values of 2 and 5, we introduce further control over the tree's growth, helping avoid unnecessary complexity. Together, these parameters aim to strike a balance between model performance and overfitting.

````python
# Grid search with five-fold cross validation

start_time = time.time() # Start timer

grid_search = GridSearchCV(dt_pipeline, param_grid, cv=5, scoring='f1', n_jobs=2)
grid_search.fit(X_train_final, y_train)

end_time = time.time()
runtime = end_time - start_time
print(f"\nGridSearchCV completed in {runtime:.2f} seconds.") # Print runtime
````

To tune our decision tree classifier, we use GridSearchCV() with five-fold cross-validation. This allows us to systematically evaluate all combinations of the hyperparameters in our grid using the F1 score as our performance metric, which balances precision and recall and is ideal for our binary classification task. We parallelize the computation with n_jobs=2 to speed up the search.

The most recent run of the grid search took 975 seconds, or just over 16 minutes. Given the size of our dataset and the computational demands of training decision trees on multiple cross-validation splits, we intentionally keep the parameter grid compact. Expanding the grid to include additional values or hyperparameters would substantially increase runtime and may not yield proportionately better performance. Our setup strikes a practical balance between thorough tuning and computational feasibility.

## 3.3 Analysis of Most Predictive Model

We now turn to a more detailed evaluation of our best-performing model by analyzing the confusion matrix and key performance metrics such as precision, recall, and F1 score, in order to better understand its strengths and limitations.

````python
# Define function to display confusion matrix

def conf_matrix(y_test,y_pred,model):
    
    """
    Given a model and predicted values, displays confusion matrix
    """
    
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)

    fig, ax = plt.subplots()
    disp.plot(cmap='Blues', ax=ax, values_format='d')  # values_format = 'd'  avoids scientific notation
    plt.title("Confusion Matrix")
    plt.show()
    
````

````python
# Define a function to examine reviews after they have been classified

def examine_classified_reviews(pred_value, true_value, y_test, y_pred, df = og_hotel_df, n_examples=5):
    
    """
    Examine example reviews based on prediction outcome.
    """

    # Ensure full review is shown
    pd.set_option('display.max_colwidth', None)

    # Determine the classification case
    if pred_value == 1 and true_value == 1:
        label = "True Positive Examples"
    elif pred_value == 1 and true_value == 0:
        label = "False Positive Examples"
    elif pred_value == 0 and true_value == 1:
        label = "False Negative Examples"
    elif pred_value == 0 and true_value == 0:
        label = "True Negative Examples"

    print(f"\n Examining: {label}\n")

    # Filter for matching classification outcome
    test_indices = y_test.index
    results_df = pd.DataFrame({
        'true_label': y_test,
        'predicted_label': y_pred
    }, index=test_indices)

    merged_df = df.loc[test_indices].copy()
    merged_df['true_label'] = results_df['true_label']
    merged_df['predicted_label'] = results_df['predicted_label']

    subset = merged_df[
        (merged_df['true_label'] == true_value) & 
        (merged_df['predicted_label'] == pred_value)
    ]

    # Sample and print reviews
    for i in range(n_examples):
        row = subset.sample(n = 1,random_state = i)
        print(f"Example {i + 1}")
        print("Positive Review:\n", row['Positive_Review'].iloc[0])
        print("\nNegative Review:\n", row['Negative_Review'].iloc[0])
        print('='*80 + '\n')
````
